```python
import pandas as pd
```

```python
data = pd.read_csv("path/train.csv",encoding='latin-1')
data.head(5)
```

```python
import nltk
import re
```

```python
# Removing digits
def digit_remove(text):
    return re.sub("\d", " ", text)
```

```python
data['digit_remove'] = data['SentimentText'].apply(digit_remove)
data.head(5)
```

```python
#Removing URLs
def remove_url(text):
    return re.sub(r'http\S+','',text)   
```

```python
data['URL_Removed'] = data['digit_remove'].apply(remove_url)
data.head(5)
```

```python
#Removing Firstword
def firstword(text):
    return re.sub(r'^[@$]\w+','',text)
```

```python
data['firstname'] = data['URL_Removed'].apply(firstword)
data.loc[[15]]
```

```python
# Eliminating other symbols
def eliminatesymbols(text):
    return re.sub("[^0-9a-zA-Z ]"," ",text)
```

```python
data['eliminatesymbols'] = data['firstname'].apply(eliminatesymbols)
data.loc[[15]]
```

```python
#Removing Usernames
def username(aao):
    a = ''.join(re.findall(r'^[@|&]\w+',aao))
    result=re.sub(a,'',aao)     
    return result
```

```python
data['username'] = data['eliminatesymbols'].apply(username)
data.head(5)
data.iloc[[15]]
```

```python
def get_word_features(wordlist):
    wordlist = nltk.FreqDist(wordlist)
    word_features = wordlist.keys()
    return word_features
```

```python
def get_words_in_tweets(tweets):
    all_words = []
    for (words, sentiment) in tweets:
      all_words.extend(words)
    return all_words
```

```python
# extract the word features out from the training data
word_features = get_word_features(get_words_in_tweets(tweets))
word_features
```

```python
def extract_features(document):
    document_words = set(document)
    features = {}
    for word in word_features:
      features['contains(%s)' % word] = (word in document_words)
    return features
```
